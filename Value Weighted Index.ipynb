```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
crsp=pd.read_csv('crsp.csv')
crsp.head()
```

```python
share_codes=[10,11]
exch_codes=[1,2,3]
crsp_codes=crsp[crsp.SHRCD.isin(share_codes) & crsp.EXCHCD.isin(exch_codes)]
crsp_codes.head()
```

```python
crsp_codes.loc[:,'mktcap']=abs(crsp_codes['PRC'])*crsp_codes['SHROUT']
crsp_codes.head()
```

```python
crsp_codes.loc[:,'date']=pd.to_datetime(crsp_codes.date,format="%Y%m%d")
crsp_codes.head()
```

```python
crsp_codes=crsp_codes.sort_values(by=['PERMNO','date'])
crsp_codes.head()
```

```python
crsp_codes.loc[:,'mktcap_lag']=crsp_codes.mktcap.shift(1)
crsp_codes.head()
```

```python
crsp_codes['RET']=pd.to_numeric(crsp_codes['RET'],errors='coerce')
crsp_codes['DLRET']=pd.to_numeric(crsp_codes['DLRET'],errors='coerce')
crsp_codes['cumret']=crsp_codes['RET']
crsp_codes.loc[crsp_codes.RET.isnull(),'cumret']=crsp_codes.DLRET
crsp_codes.loc[crsp_codes.DLRET.notnull() & crsp_codes.RET.notnull(),'cumret']=(1+crsp_codes.DLRET)*(1+crsp_codes.RET)-1
crsp_codes.head()
```

```python
crsp_filter=crsp_codes.iloc[:,[0,1,8,9,10]]
crsp_filter.head()
```

```python
lag_mktcap=pd.DataFrame(crsp_filter.groupby('date')['mktcap_lag'].sum())
lag_mktcap.columns=['total_mkt_cap']
lag_mktcap
```

```python
crsp_filter=crsp_filter.merge(lag_mktcap,left_on='date',right_index=True)
crsp_filter=crsp_filter.sort_values(by=['PERMNO','date'])
crsp_filter.head()
```

```python
crsp_filter.loc[:,'weight']=crsp_filter.mktcap_lag/crsp_filter.total_mkt_cap
crsp_filter.loc[:,'vwretd']=crsp_filter.weight*crsp_filter.cumret
crsp_filter.head()
```

```python
value_returns=pd.DataFrame(crsp_filter.groupby(['date'])['vwretd'].sum())
value_returns=value_returns.reset_index()
value_returns['month']=value_returns.date.dt.month
value_returns['year']=value_returns.date.dt.year
value_returns.head()
```

```python
equal_returns=pd.DataFrame(crsp_filter.groupby('date')['cumret'].mean())
equal_returns=equal_returns.reset_index()
equal_returns['month']=equal_returns.date.dt.month
equal_returns['year']=equal_returns.date.dt.year
equal_returns.head()
```

```python
## FF Factors
ff_values=pd.read_csv('F-F_Research_Data_Factors.csv')
ff_values.head()
```

```python
ff_values['RF']=ff_values.RF/100
ff_values['date']=pd.to_datetime(ff_values['Unnamed: 0'],format='%Y%m')
ff_values['month']=ff_values.date.dt.month
ff_values['year']=ff_values.date.dt.year
ff_values.head()
```

```python
rf_rate=ff_values.iloc[:,4:8]
rf_rate.head()
```

```python
all_returns=value_returns.merge(equal_returns,on=['month','year']).merge(rf_rate,on=['month','year'])
all_returns=all_returns.iloc[:,[2,3,1,5,6]]
all_returns=all_returns.rename(columns={'cumret':'ewretd'})
all_returns.head()
```

```python
all_returns['e_vwretd']=all_returns.vwretd-all_returns.RF
all_returns['e_ewretd']=all_returns.ewretd-all_returns.RF
all_returns
```

```python
from statistics import stdev
from math import sqrt
value_mean=all_returns.e_vwretd.mean()*12
value_sd=stdev(all_returns.e_vwretd)*sqrt(12)
value_sr=value_mean/value_sd
print(value_mean*100)
print(value_sd*100)
print(value_sr)
```

```python
ff_returns=ff_values['Mkt-RF']-ff_values['RF']
print(ff_returns.mean()*12)
print(stdev(ff_returns)*sqrt(12))
print(ff_returns.mean()/stdev(ff_returns)*sqrt(12))
```

```python
ff_values['e_ff_ret']=ff_values['Mkt-RF']-ff_values['RF']
ff_vwretd=ff_values.iloc[:,6:9]
ff_vwretd
```

```python
corr_data=all_returns.loc[:,['month','year','e_vwretd']].merge(ff_vwretd,on=['month','year'])
corr_data['e_ff_ret']=corr_data['e_ff_ret']/100
corr_data
```

```python
corr_data.e_vwretd.corr(corr_data.e_ff_ret)
```

```python
#Maximum Absolute Difference

corr_data['mad']=abs(corr_data.e_vwretd-corr_data.e_ff_ret)
max(corr_data['mad'])
```
